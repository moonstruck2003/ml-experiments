```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import sklearn.datasets 
from sklearn.model_selection import train_test_split
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer() 
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
x = data_frame.drop(columns='label',axis=1)
y = data_frame['label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
#Standardized the data 
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
x_train_std = scaler.fit_transform(x_train) #only for training data, for test data only transfor 
x_test_std = scaler.transform(x_test)
```

```python
#Building the Neural Network 

import tensorflow as tf 
tf.random.set_seed(3) #how many time you run your code you will get the same accuracy score  
from tensorflow import keras 
```

```python
#Setting up the layers of my neural network 
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)) , #input = no of features (convert the data into single dim array
    keras.layers.Dense(20, activation='relu') , # hidden 
    keras.layers.Dense(2,activation='sigmoid')  # output , no of classes . here only 2. 
])
```

```python
#compiling the neural network 
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# dog 0 cat 1 horse 2 ----> sparse_catagorical_crossentropy 
# dog 0 0 1 cat 0 1 0 horse 0 0 1 ---> catagorical_crossentropy 
```

```python
#training the neural network 
history = model.fit(x_train_std,y_train, validation_split=0.1, epochs=10)
```

```python
#Visualizing the accuracy and loss 
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy']) 

plt.title('Model Accuracy ')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
#Visualizing the accuracy and loss 
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_loss']) 

plt.title('Model loss ')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
#Accurary of the model on test data 
loss,accuracy = model.evaluate(x_test_std,y_test)
print(accuracy)
```
