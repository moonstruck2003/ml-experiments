```python
#correlation values of GLD
print(correlation['GLD'])
```

```python
#check the distribution of gold -price
sns.distplot(gold_data['GLD'],color='green')
```

```python
#SPITTING THE FEATURES AND TARGETS
x = gold_data.drop(['Date','GLD'],axis=1)
y= gold_data['GLD']
```

```python
#splitting into training data and test data
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Model Training Random Forest Regressor 
reg = RandomForestRegressor(n_estimators=100)
#trianing the model
reg.fit(x_train,y_train)
```

```python
#model evaluation
#prediction on test data
test_data_pre = reg.predict(x_test)
print(test_data_pre)
```

```python
#r squared error
error_score = metrics.r2_score(y_test,test_data_pre)
print(error_score)
```

```python
#compare the actual values and predicted values in plot

y_test = list(y_test)
```

```python
plt.plot(y_test, color='blue', marker='o', linestyle='-', label='Actual Value')
plt.plot(test_data_pre, color='green', linestyle='-', label='Predicted Value')
plt.title("Actual Price vs Predicted Price")
plt.xlabel("Number of Values")
plt.ylabel("Gold Price")
plt.legend()
plt.show()
```
