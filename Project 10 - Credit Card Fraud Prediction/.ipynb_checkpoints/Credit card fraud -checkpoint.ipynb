```python
# dataset link =  "https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud"
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
file_path = r"C:\Users\PC\Downloads\cod\Compressed\archive_2\creditcard.csv"
credit_data = pd.read_csv(file_path)
credit_data.head(2)
```

```python
credit_data.shape
```

```python
credit_data.info
```

```python
credit_data.isnull().sum()
```

```python
credit_data['Class'].value_counts()
```

```python
#seperating the 0 and 1s
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#under-sampling
legit_sample = legit.sample(n=492)
legit_sample.head(2)
```

```python
#concating the legit_sample and fraud columns
new_data = pd.concat([legit_sample,fraud],axis=0)
new_data.head()
```

```python
new_data.shape
```

```python
new_data['Class'].value_counts()
```

```python
#seperating the labels and features
x = new_data.drop(columns='Class',axis=1)
y = new_data['Class']
```

```python
print(x)
```

```python
#Standardization of data
scaler = StandardScaler()
scaler.fit(x)
```

```python
x = scaler.transform(x)
```

```python
#train test split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
#model training
log = LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
#accuracy test
train_data_pre = log.predict(x_train)
train_data_accuraccy = accuracy_score(train_data_pre,y_train)
print(train_data_accuraccy)
```

```python
#test data accuraccy
test_data_prediction = log.predict(x_test)
test_data_accuracy = accuracy_score(test_data_prediction,y_test)
print(test_data_accuracy)
```
